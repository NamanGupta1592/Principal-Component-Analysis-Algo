```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
X=wine.data
y = wine.target
```

```python
x= pd.DataFrame(X,columns=wine.feature_names)
```

```python
y = pd.DataFrame(y,columns=['target'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_reduced=pca.fit_transform(x)
```

```python
x.shape
```

```python
x_reduced.shape
```

```python
pca.components_
```

```python
pca.components_.shape
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
# finding  the right number of components by okotting the no of components Vs explained variance ratio
from sklearn.decomposition import PCA
pca2 = PCA()
pca2.fit(x_scaled)

exp_var_ratio = pca2.explained_variance_ratio_
cumsum = np.cumsum(exp_var_ratio)
plt.plot(cumsum)
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.show()
```

```python
#finding the number of components from desired explained variance ratio
from sklearn.decomposition import PCA
pca3 = PCA(0.96)
x_reduced3=pca3.fit_transform(x_scaled)
```
